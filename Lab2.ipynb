Processen att försöka upptäcka meningsfulla mönster och samband i ostrukturerad information/ostrukurerade data dvs. texter kallas för textmining. Text mining är ett väldigt omdiskuterat område inom området "data analytics", vilket kan sägas vara en delmängd av del av området artificiell intelligens (AI).  Två väldigt viktiga applikationer av textmining är "sentiments analysis" och "web mining".  Den förstnämnda applikationen innebär att det görs en analys av de korta texter som förmedlas över sociala nätverk. Den här typen av analyser görs i regel på texter med ett begränsat antal tecken. Den här typen av analyser används för att analysera attityder, utvärderingar, opinioner och känslorna hos användare gällare olika objekt, händelser, problem, produkter, tjänster och olika områden. De har använts framgångsrikt vid marknadsföring av nya produkter, predicera matcher och upptäcka trender vid röstningskampanjer [1]. Detta är bakgrunden till varför jag valde en liknande lösning för laboration 2, samt att jag hade tillgång till material som gjorde detta möjligt [2,3,4,5,6]. Den lösning som jag presenterar i laboration är avgränsad då tiden och kunskapen är begränsad och att laborationen görs på en grundläggande kurs inom området.

Det finns olika sätt att representera en text som en vektor och sedan utföra en "sentiment" klassificering på dessa representationer. En nackdel med ett sådant angreppssätt är att man då inte tar hänsyn i vilken ordning som orden kommer. Recurrent Neural Networks (RNN) är lämpliga att använda när det är viktigt att behålla ordningen på orden. RNN:s är användbara när man vill predicera nästa händelse givet en sekvens av händelser och anses vara ett sekvensiellt problem inom området Natural Language Processing (NLP). "Sentiment analysis" är ett utmanande problem även för människor för ord har olika mening i olika kontexter. RNN:s anses vara den bästa formen av neurala nätverk för sekvensiella problem. De flesta av denna typ av problem är klassificeringsproblem (positiva/neutrala/negativa) och inte regressionsproblem. RNN:s är neurala nätverk med flera lager. Indata delas upp i olika tidssteg, t.ex ifall indata är en mening så delas meningen upp i olika (i detta fallet representerar varje ett tidssteg). Varje ord fungerar sedan som indata till nästa lager. Den utdata från föregående steg som blir till indata i kommande steg kallas för det gömda tillståndet (hidden state). Detta tillstånd tillåter ett RNN att ha ett "minne" över de olika tillstånden i den sekvensiella indatan. Det sk. "vanishing gradient problemet" får dock till följd att RNN:s inte har möjligheten att lära sig från tidigare lager (tidiga tidssteg) vilket får till följd att långtidsminnet blir sämre [3,4,5,6].

Datasetet innehåller tweets om flygbolaget US Airline från februari år 2015 som klassificerats i positiva, negativa och neutrala tweets. De negativa tweetsen klassificeraras också på basis av orsak. Målet med uppgiften är att bygga en modell som predicerar ifall en tweets författares uppfattning är positiv, neutral eller negativ [2].

Jag använder mig i uppgiften både av matplotlib såväl som TensorBoard då jag ville lära mig mer om det sistnämnda. För att se matplotdiagrammen så hänvisas till längst ner i programkoden nedan. Användningen av hyperparametern "padding". Textrecensionerna som används i denna uppgiften har olika längd så därför är vektorerna som representerar indata av olika längd. Detta är ett centralt problem då neurala nätverk bara accepterar vektorer med en viss storlek. För att addressera detta problem så används maxlen parametern som är den maximala längden av varje textrecension. Recensioner som är längre än maxlen blir trunkerade och recensioner som är kortare än fylls ut med nollor dvs. "padding".  

Noggrannheten (accuracy) för träning och validering presenteras för modellen som tränats med hjälp av opimeraren adam. Detta gapet på 11 % anger att en viss överträning (overfitting) sker när optimeraren adam används. Det är också värt att notera att SimpleRNN inte presterar särskilt bra på längre sekvenser [6].  Noggrannheten hos modellen kan säkert förbättras om antalet datapunkter ökas. Andra åtgärder som kan vidtas för att förbättra prestanda är bl.a .att förbättra förbearbetningen av text (bättre rensning av text),  optimera hyperparametrarna och även betänka att den ursprungliga etiketteringen (labels) kanske inte varit optimal utan varit obalanserad i någon riktning, positiv, negativ eller neutral (det krävs dock en noggrannare analys för att egentligen kunna uttala sig om detta).

```python
# Hämtar datasetet "Twitter US Airline Sentiment"
# Datasetet innehåller Twitter data som samlades in från februari år 2015 från resenärer hos flygbolaget US Airline 
# Resenärerna fick först klassificera positiva, negativa och neutrala tweets gällande ett flygbolag för att sedan kategorisera negativa orsaker

!wget https://www.dropbox.com/s/8yq0edd4q908xqw/airline_sentiment.csv?dl=1
```

```python
# Laddar och inspekterar datasetet

t=pd.read_csv('/content/airline_sentiment.csv?dl=1')
t.head()
```

```python
# Förbearbeta texten för att ta bort punktation, normalisera all ord till gemener och ta bort stoppord dvs betydelsefattiga ord (pre-processing) 

nltk.download('stopwords')
stop = nltk.corpus.stopwords.words('english')
def preprocess(text):
    text=text.lower()
    text=re.sub('[^0-9a-zA-Z]+',' ',text)
    words = text.split()
    words2=[w for w in words if (w not in stop)]
    #words3=[ps.stem(w) for w in words]
    words4=' '.join(words2)
    return(words4)
t['text'] = t['text'].apply(preprocess)
```

```python
# Extrahera mappningen av ord till heltal för alla de ord som datasetet består av 

from collections import Counter
counts = Counter()
for i,review in enumerate(t['text']):
    counts.update(review.split())
```

```python
# Extrahera frekvensen av alla ord i datasetet och presentera några exempel

words = sorted(counts, key=counts.get, reverse=True)
words[0:10]
```

```python
# En loop används för att gå igenom alla ord, tildela varje ord ett index och presenterar det

word_to_int = {word: i for i, word in enumerate(words, 1)}
int_to_word = {i: word for i, word in enumerate(words, 1)}
int_to_word
```

```python
# Mappning av varje ord i en mening till motsvarande ord som är associerat med ordet

mapped_reviews = []
for review in t['text']:
    mapped_reviews.append([word_to_int[word] for word in review.split()])
```

```python
# I det föregående steget konverteras en textrecension till en lista av listor där varje lista identifierar orden i meningen och här presenteras ett exempel på original och mappning 

print('Original text:',t.loc[0]['text'])
print('Mapped text:',mapped_reviews[0])
```

```python
# Extrahera den maximala längden av en mening och normalisera alla meningar till samma längd genom "padding"
# En loop används för att gå igenom alla textrecensioner och spara längden på varje recension
# Slutligen beräknas den maximala längden på en textrecension

length_sent = []
for i in range(len(mapped_reviews)):
  length_sent.append(len(mapped_reviews[i]))
sequence_length = max(length_sent)
```

```python
# Tweets är av olika längd men RNN:s förväntar sig att antalet steg för varje indata att vara densamma
# I nedan görs en "padding" av en mappad textrecension med värdet 0, ifall längden på textrecensionen är mindre än den maximala längden hos alla textrecensionerna i datasetet

X = pad_sequences(maxlen=sequence_length, sequences=mapped_reviews, padding="post", value=0)
```

```python
# Förbereder tränings- och testdataset. Originaldata delas in i tränings- och testdata 
# Den beroende variabeln konverterar till en "one-hot-encoded" variabel. "One-hot-encoded" är en process i vilken kategoriska variabler konverteras till en form som är lättare för en maskinlärningsalgoritm att använda vid prediktion

y=t['airline_sentiment'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=10)
y_train2 = to_categorical(y_train)
y_test2 = to_categorical(y_test)
```

```python
# Bygger RNN arkitekturen och kompilerar modellen
# Embedding tar det totala antalet distinkta ord tas som indata och skapar en vektor för varje ord där output_dim representerar antalet dimensioner som ordet skall representeras i och input_length representerar antalet ord i varje mening
# Notera att om vi i RNN lagret vill extrahera utdata för varje steg så är return_sequence lika med True men i det här fallet så extraheras utdata enbat efter att ha läst igenom alla ord som är indata och sedan returnerar return_squences är lika med False

embedding_vecor_length=32
max_review_length=26
model = Sequential()
model.add(Embedding(input_dim=12533, output_dim=32, input_length = 26))
model.add(SimpleRNN(40, return_sequences=False))
model.add(Dense(2, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# En summering av modellen presenteras

print (model.summary())
```

```python
# Keras skriver TensorBoard log filer genom sk "callbacks", vilka möjliggör en visualisering av dynamiska grafer av tränings- och testmätetal (metrics), såväl som aktiveringshistogram för olika lager av modellen

tbCallBack = TensorBoard(log_dir='./log', histogram_freq=1,
                         write_graph=True,
                         write_grads=True,
                         batch_size=32,
                         write_images=True)
```

```python
# Tränar (fit) modellen. Medan modellen körs så går det att följa förloppet genom att öppna ngrok länken som presenterats ovan

history = model.fit(X_train, y_train2, validation_data=(X_test, y_test2), epochs=10, batch_size=32,callbacks=[tbCallBack])
```

```python
# Noggrannheten (accuracy) och "loss" värden i tränings- och testdataset presenteras med hjälp av matplotplib

history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
acc_values = history_dict['acc']
val_acc_values = history_dict['val_acc']
epochs = range(1, len(val_loss_values) + 1)
import matplotlib.ticker as mtick
plt.subplot(211)
plt.plot(epochs, history.history['loss'], 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'r', label='Test loss')
plt.title('Training and test loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid('off')
plt.show()
plt.subplot(212)
plt.plot(epochs, history.history['acc'], 'bo', label='Training accuracy')
plt.plot(epochs, val_acc_values, 'r', label='Test accuracy')
plt.title('Training and test accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
plt.legend()
plt.grid('off')
plt.show()
```
